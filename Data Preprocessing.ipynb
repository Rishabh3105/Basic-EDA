## Data exploration using Data Visualisation

```python
import matplotlib.pyplot as plt
```

```python
fig1,ax=plt.subplots(2,2,figsize=(10,10))
ax[0][0].hist(df.age)
ax[0][0].set_ylabel("Frequency")
ax[0][0].set_xlabel("Age")
ax[0][1].hist(df.sim)
ax[0][1].set_ylabel("Frequency")
ax[0][1].set_xlabel("Sim")
ax[1][0].hist(df.gender)
ax[1][0].set_ylabel("Frequency")
ax[1][0].set_xlabel("Gender")
ax[1][1].hist(df.complaint)
ax[1][1].set_ylabel("Frequency")
ax[1][1].set_xlabel("Complaint")

```

### Fig1.  Frequency of age lying in category 1 is highest
### Fig2.  Frequency of Single sim phones is too high compared to dual sim
### Fig3.  Data mainly consists of Gender == "Male"
### Fig4.  Represents the frequency of compalints count 

```python
fig1,ax=plt.subplots(2,2,figsize=(10,10))
ax[0][0].boxplot(df.sms, vert=False)
ax[0][0].set_xlabel("Frequency")
ax[0][0].set_ylabel("SMS")
ax[0][1].boxplot(df.mms, vert=False)
ax[0][1].set_xlabel("Frequency")
ax[0][1].set_ylabel("MMS")
ax[1][0].boxplot(df.charges, vert=False)
ax[1][0].set_xlabel("Frequency")
ax[1][0].set_ylabel("Charges")
ax[1][1].boxplot(df.complaint, vert=False)
ax[1][1].set_xlabel("Frequency")
ax[1][1].set_ylabel("Complaint")
```

```python
a_mean=df.charges.mean()
a_max= df.charges.max()
a_min= df.charges.min()
a_median= df.charges.median()
print("mean is {}\nmax is {}\nmin is {}\nmedian is {}\n".format(a_mean,a_max,a_min,a_median))

fig1,ax=plt.subplots(2,1,figsize=(10,10))
ax[0].hist(df.charges)
ax[0].set_ylabel("Frequency")
ax[0].set_xlabel("Charges")
ax[0].axvline(x=a_min,color="grey", linestyle='dashed',linewidth=3)
ax[0].axvline(x=a_max,color="grey", linestyle='dashed',linewidth=3)
ax[0].axvline(x=a_mean,color="red", linestyle='dashed',linewidth=3)
ax[0].axvline(x=a_median,color="pink", linestyle='dashed',linewidth=3)
ax[1].boxplot(df.charges, vert=False)
ax[1].set_xlabel("Frequency")
ax[1].set_ylabel("Charges")
ax[1].axvline(x=a_min,color="grey", linestyle='dashed',linewidth=3)
ax[1].axvline(x=a_max,color="grey", linestyle='dashed',linewidth=3)
ax[1].axvline(x=a_mean,color="red", linestyle='dashed',linewidth=3)
ax[1].axvline(x=a_median,color="pink", linestyle='dashed',linewidth=3)
```
